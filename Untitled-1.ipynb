```python
import cv2
import pytesseract
import pandas as pd

# Load the video file
video_path = 'engagement.mp4'
video = cv2.VideoCapture(video_path)

# Get video properties
fps = video.get(cv2.CAP_PROP_FPS)
frame_skip = int(fps * 5)  # Process every 5 seconds

pytesseract.pytesseract.tesseract_cmd = '/opt/homebrew/bin/tesseract'  # Update path for Tesseract

# Function to preprocess and extract numbers from a frame
def extract_number_from_frame(frame):
    # Convert to grayscale
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Apply Gaussian blur to reduce noise
    blurred = cv2.GaussianBlur(gray_frame, (5, 5), 0)

    # Apply thresholding to make the text more clear
    _, thresh = cv2.threshold(blurred, 150, 255, cv2.THRESH_BINARY_INV)

    # Optional: Resize the image if the text is too small
    resized = cv2.resize(thresh, None, fx=2, fy=2, interpolation=cv2.INTER_LINEAR)

    # Use Tesseract to extract the text with more specific settings
    config = '--psm 7'  # PSM 7 is good for single line of text
    text = pytesseract.image_to_string(resized, config=config)

    # Look for numeric values in the extracted text
    for line in text.split("\n"):
        line = line.strip()
        if line.isdigit():  # Ensure the text is a valid number
            return int(line)
    return None

# Prepare to store the data
data = []
frame_idx = 0

# Process the video
success, frame = video.read()
while success:
    number = extract_number_from_frame(frame)
    Timestepstamp = frame_idx / fps  # Calculate the timestamp

    if number is not None:
        data.append({"Timestamp": timestamp, "Value": number})
    else:
        # If no number detected, store a placeholder (-1)
        data.append({"Timestamp": timestamp, "Value": -1})

    # Skip to the next frame after the specified interval (every 5 seconds)
    frame_idx += frame_skip
    video.set(cv2.CAP_PROP_POS_FRAMES, frame_idx)
    success, frame = video.read()

# Save the data to CSV
df = pd.DataFrame(data)
df.to_csv('engagement.csv', index=False)
print("Data saved to engagement.csv")
```

```python
#  EMOTIV provides six basic measures of mental performance, derived directly from your mental activity. Each measure is automatically scaled to suit your normal range and base level of each condition – the system learns your usual state and capabilities and provides an adjusted value showing your relative performance on each occasion, compared to your overall behavior. 

# Performance metrics graphs include 6 metrics for INSIGHT, EPOC X and EPOC+ headsets:

# Stress (FRU) is a measure of comfort with the current challenge. High stress can result from an inability to complete a difficult task, feeling overwhelmed and fearing negative consequences for failing to satisfy the task requirements. Generally a low to moderate level of stress can improve productivity, whereas a higher level tends to be destructive and can have long term consequences for health and wellbeing.
# Engagement (ENG) is experienced as alertness and the conscious direction of attention towards task-relevant stimuli. It measures the level of immersion in the moment and is a mixture of attention and concentration and contrasts with boredom. Engagement is characterized by increased physiological arousal and beta waves along with attenuated alpha waves. The greater the attention, focus, and workload, the greater the output score reported by the detection.
# Interest (VAL) is the degree of attraction or aversion to the current stimuli, environment or activity and is commonly referred to as Valence. Low interest scores indicate a strong aversion to the task, high interest indicates a strong affinity with the task while mid-range scores indicate you neither like nor dislike the activity.
# Excitement (EXC) is an awareness or feeling of physiological arousal with a positive value. It is characterized by activation in the sympathetic nervous system which results in a range of physiological responses including pupil dilation, eye widening, sweat gland stimulation, heart rate and muscle tension increases, blood diversion, and digestive inhibition. In general, the greater the increase in physiological arousal the greater the output score for the detection. The Excitement detection is tuned to provide output scores that reflect short-term changes in excitement over time periods as short as several seconds.
# Focus (FOC) is a measure of fixed attention to one specific task. Focus measures the depth of attention as well as the frequency that attention switches between tasks. A high level of task switching is an indication of poor focus and distraction.
# Relaxation (MED) is a measure of an ability to switch off and recover from intense concentration. Trained meditators can score extremely high relaxation scores.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter

# Replace 'run1.csv' with the path to your CSV file
df = pd.read_csv('run1.csv')

# Apply Savitzky-Golay filter to smooth the lines
df['Attention_smoothed'] = savgol_filter(df['Attention'], window_length=5, polyorder=2)
df['Engagement_smoothed'] = savgol_filter(df['Engagement'], window_length=5, polyorder=2)
df['Stress_smoothed'] = savgol_filter(df['Stress'], window_length=5, polyorder=2)

# Define task durations with start and end times
tasks = [
    {"Start": 1730353650, "End": 1730353830, "Description": "Task 1"},
    {"Start": 1730353860, "End": 1730353980, "Description": "Task 2"},
    {"Start": 1730354040, "End": 1730354190, "Description": "Task 3"}
]

# Calculate average values for each metric within each task
task_names = [task["Description"] for task in tasks]
avg_attention = []
avg_engagement = []
avg_stress = []

for task in tasks:
    # Filter data within the task duration
    task_data = df[(df['Time'] >= task["Start"]) & (df['Time'] <= task["End"])]
    
    # Append average values for each metric during the task
    avg_attention.append(task_data['Attention_smoothed'].mean())
    avg_engagement.append(task_data['Engagement_smoothed'].mean())
    avg_stress.append(task_data['Stress_smoothed'].mean())

# Plotting three separate bar graphs
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)

# Attention Bar Plot
axes[0].bar(task_names, avg_attention, color='blue')
axes[0].set_title('Average Attention per Task')
axes[0].set_xlabel('Tasks')
axes[0].set_ylabel('Average Value')

# Engagement Bar Plot
axes[1].bar(task_names, avg_engagement, color='orange')
axes[1].set_title('Average Engagement per Task')
axes[1].set_xlabel('Tasks')

# Stress Bar Plot
axes[2].bar(task_names, avg_stress, color='green')
axes[2].set_title('Average Stress per Task')
axes[2].set_xlabel('Tasks')

# Show the plot
plt.suptitle('Average Attention, Engagement, and Stress per Task')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust layout to fit title
plt.show()
```
